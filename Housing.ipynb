```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("Housing.csv")
```

```python
data_set
```

```python
data_set["mainroad"] = data_set["mainroad"].astype("category")
data_set["guestroom"] = data_set["guestroom"].astype("category")
data_set["basement"] = data_set["basement"].astype("category")
data_set["hotwaterheating"] = data_set["hotwaterheating"].astype("category")
data_set["airconditioning"] = data_set["airconditioning"].astype("category")
data_set["prefarea"] = data_set["prefarea"].astype("category")
data_set["furnishingstatus"] = data_set["furnishingstatus"].astype("category")
```

```python
obejects = data_set.drop(['price','area','bedrooms','bathrooms','stories','parking'], axis = 1)
```

```python
obejects
```

```python
obejects["mainroad"] = obejects["mainroad"].cat.codes
obejects["guestroom"] = obejects["guestroom"].cat.codes
obejects["basement"] = obejects["basement"].cat.codes
obejects["hotwaterheating"] = obejects["hotwaterheating"].cat.codes
obejects["airconditioning"] = obejects["airconditioning"].cat.codes
obejects["prefarea"] = obejects["prefarea"].cat.codes
obejects["furnishingstatus"] = obejects["furnishingstatus"].cat.codes
```

```python
obejects
```

```python
numerical_data = data_set.drop(['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus'], axis = 1)
```

```python
numerical_data
```

```python
new_data_set = numerical_data.join(obejects)
```

```python
new_data_set
```

```python
data_values = new_data_set.values
```

```python
data_values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
new_target = data_values[:,0]
new_data = data_values[:,1:]
train_data, test_data, train_target, test_target = train_test_split(new_data, new_target, test_size = 0.1)
```

```python
model = LinearRegression()
model.fit(train_data, train_target)
```

```python
predict = model.predict(test_data)

from sklearn.metrics import r2_score

r2 = r2_score(test_target, predict)
print("r2 score:", r2)
```
